# Basic Data Exploration

- The dataset contains 284,807 rows and 31 columns.
- No missing values are present.
- All features are numerical.
- Target variable:
  - 0 → Normal transaction
  - 1 → Fraud transaction

The dataset is highly imbalanced, which can affect model performance.

```python
# Class Imbalance Analysis
fraud = df['Class'].value_counts()[1]
normal = df['Class'].value_counts()[0]

fraud_percentage = (fraud/(fraud+normal))*100
print(f"Fraud Percentage: {fraud_percentage:.2f}%")
```

# Class Imbalance Analysis

- Fraud cases are extremely low compared to normal transactions.
- Fraud percentage is approximately **0.17%**, indicating a highly imbalanced dataset.

# Why is this important?

This imbalance can lead to misleading model performance, where the model may predict all transactions as normal and still achieve high accuracy.

-  Therefore, accuracy alone is not a reliable metric for this problem.

```python
sns.countplot(x='Class', data=df)
plt.title("Fraud vs Normal Transactions")
plt.xlabel("Class (0 = Normal, 1 = Fraud)")
plt.ylabel("Count")
plt.show()
```

The visualization clearly shows that fraudulent transactions are extremely rare compared to normal transactions, confirming severe class imbalance.

```python
sns.boxplot(x='Class', y='Amount', data=df)
plt.title("Transaction Amount vs Class")
plt.show()
```

# Transaction Amount vs Class

This visualization compares transaction amounts between normal and fraudulent transactions.

# Observations:
- Most transactions have low amounts.
- Fraudulent transactions generally occur in lower amount ranges.
- There are some extreme outliers in normal transactions.

# Insight:
Transaction amount alone is not sufficient to detect fraud, but it can still provide useful patterns when combined with other features.

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), cmap='coolwarm')
plt.title("Feature Correlation")
plt.show()
```

# Feature Correlation Analysis

The correlation heatmap shows relationships between different features in the dataset.

# Observations:
- Most features have very low correlation with each other.
- This is because the dataset features (V1 to V28) are transformed using PCA.
- Low correlation helps reduce multicollinearity.

# Insight:
Since features are already transformed and independent, machine learning models can perform better without needing further dimensionality reduction.
